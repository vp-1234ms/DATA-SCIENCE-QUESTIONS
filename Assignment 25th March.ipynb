they may impact your analysis.

```python
df.dtypes
```

```python
sns.boxplot(df['Price'])
```

```python
#The impact of outliers on the dataset can be summarized as follows:

#Effect on measures of central tendency: Outliers can significantly affect measures of central tendency, such as the mean and median. 
#The mean is particularly sensitive to outliers, as it takes into account the values of all data points, while the median is
#less affected as it only considers the middle value(s). Therefore, outliers can skew the mean, making it larger or smaller than the median.

#Effect on measures of spread: Outliers can also affect measures of spread, such as the standard deviation and range. 
#Outliers can increase the variance and standard deviation of the data, making it appear more variable than it actually is. 
#Outliers can also widen the range, making the data appear more spread out.

#Effect on model performance: Outliers can have a significant impact on the performance of statistical models, particularly those that
#rely on assumptions of normality and homogeneity of variance. Outliers can distort the distribution of the data, violating the assumptions
#of normality and homogeneity of variance, and can lead to biased parameter estimates, reduced model accuracy, and poor predictive performance.
```

Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
impact your analysis.

```python
df.isnull().sum()
```

```python
#Missing values in a dataset can have a significant impact on analysis in several ways. Some of the impacts are:

#Reduced sample size: Missing values reduce the number of observations available for analysis. This can result in a smaller sample size, which
#can reduce the statistical power of the analysis and make it more difficult to detect significant effects.

#Biased results: If missing values are not handled properly, they can bias the results of the analysis. For example, if missing values are more 
#likely to occur in certain groups, this can lead to biased estimates of the mean or other statistical parameters for those groups.

#Inaccurate models: Missing values can also affect the accuracy of predictive models. If missing values are not accounted for,
#models may be trained on incomplete data, leading to less accurate predictions.

#False conclusions: Missing values can also lead to false conclusions in some cases. For example, if missing values are not taken into account,
#it may appear that two variables are not related when they actually are, or that a relationship between variables exists when it does not.
```
