Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

```python
#Pandas series can be treated as list
#Where each element has it's inbuilt index starting from 0 to len(list-1)
#Whereas Dataframe is collection of such multiple series in Pandas
#Therefore we can surely say that list<series<dataframe
```

```python
#Series example
lst=[i*i for i in range(1,5)]
srs=pd.Series(lst)
srs
```

```python
#Dataframe example
lst0=[i for i in range(1,5)]
lst1=[i*i for i in range(1,5)]
lst2=[i*i*i for i in range(1,5)]
df=pd.DataFrame({"List0":lst0,"List1":lst1,"List2":lst2})
df
```
