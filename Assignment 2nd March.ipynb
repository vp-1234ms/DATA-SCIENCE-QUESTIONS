Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y label")
plt.title("This is title")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
#Subplot fucntion of matplotlib is used to display multiple plots
```

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(4,1,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(4,1,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(4,1,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(4,1,4)
plt.plot(x,y)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
#Bar plot is used to show plot categorical values or where we have some name or string on xaxis and respective value of yaxis
```

```python
plt.bar(company,profit)
plt.xlabel("This are companies")
plt.ylabel("This are profits")
plt.title("This is my graph")
```
