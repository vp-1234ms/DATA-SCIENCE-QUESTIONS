```python
#The wine quality data set contains information on various physicochemical properties of different types of wines, along with their associated 
#quality ratings. Here are the key features of the data set and their importance in predicting the quality of wine:

#Fixed acidity: This refers to the amount of acids, such as tartaric acid, that are present in the wine. It is an important feature as it can 
#impact the perceived taste and balance of the wine. Wines with too high or too low fixed acidity levels can be perceived as either too tart or 
#too flat.

#Volatile acidity: This feature measures the amount of volatile acids, such as acetic acid, in the wine. High levels of volatile acidity can make 
#the wine smell like vinegar, which is not desirable.

#Citric acid: This is a type of acid that is found in citrus fruits and can contribute to the overall balance of the wine. It is important to 
#have the right amount of citric acid in the wine to avoid it being too sweet or too sour.

#Residual sugar: This feature refers to the amount of sugar that remains in the wine after fermentation. It is an important factor in determining 
#the sweetness of the wine and can impact its perceived quality.

#Chlorides: This feature measures the amount of salt in the wine. It can contribute to the wine's overall flavor profile, but too much can make
#it taste salty or briny.

#Free sulfur dioxide: This feature measures the amount of sulfur dioxide, which is used as a preservative, that is present in the wine.
#It is important to have the right amount of free sulfur dioxide in the wine to prevent spoilage.

#Total sulfur dioxide: This feature measures the total amount of sulfur dioxide in the wine, including both free and bound sulfur dioxide.
#It is an important factor in determining the wine's stability and shelf life.

#Density: This feature measures the density of the wine, which can be influenced by the alcohol content, residual sugar, and other factors.
#It is an important factor in determining the body and mouthfeel of the wine.

#pH: This feature measures the acidity or alkalinity of the wine. It is important to have the right pH level in the wine to avoid spoilage and to 
#contribute to its overall balance and flavor profile.

#Sulphates: This feature measures the amount of sulphates, which are often added to wines as a preservative, that are present in the wine. 
#It can impact the wine's flavor and aroma profile.

#Alcohol: This feature measures the alcohol content of the wine, which can influence its body, flavor, and perceived quality. 
#Wines with too high or too low alcohol content can be perceived as unbalanced.

#All of these features are important in predicting the quality of wine as they contribute to the wine's overall flavor, aroma, balance, and stability.
#By analyzing these features and their relationships to each other, it is possible to develop predictive models that can help determine the quality 
#of different types of wine.
```

Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
Discuss the advantages and disadvantages of different imputation techniques.

```python
df.isnull().sum()
```

```python
#Missing data is a common issue in many datasets, including the wine quality dataset. It can be caused by various factors such as data entry errors,
#incomplete data collection, or data loss during transmission. In order to handle missing data in the dataset during the feature engineering 
#process, one can use different imputation techniques. These techniques involve replacing missing values with estimated values based on the 
#other available data in the dataset.

#Here are some of the most common imputation techniques and their advantages and disadvantages:

#Mean imputation: This technique involves replacing missing values with the mean value of the feature. Mean imputation is simple to implement
#and can preserve the overall distribution of the feature. However, it can also introduce bias into the dataset, especially if the feature has
#a skewed distribution.

#Median imputation: This technique involves replacing missing values with the median value of the feature. Median imputation is also simple 
#to implement and can be more robust to outliers than mean imputation. However, like mean imputation, it can also introduce bias into the dataset.

#K-nearest neighbor imputation: This technique involves finding the K nearest neighbors of the sample with missing values and using their 
#values to impute the missing data. K-nearest neighbor imputation can be more accurate than mean or median imputation as it takes into account the 
#relationships between different features in the dataset. However, it can also be computationally expensive and sensitive to the choice of K.

#Multiple imputation: This technique involves creating multiple imputations of the missing data based on a statistical model of the dataset. 
#Multiple imputation can be more accurate than single imputation techniques as it accounts for the uncertainty in the missing data. However, 
#it can also be computationally expensive and requires a good understanding of the underlying statistical model.

#Regression imputation: This technique involves using a regression model to predict the missing values based on the other features in the dataset.
#Regression imputation can be more accurate than mean or median imputation as it takes into account the relationships between different features 
#in the dataset. However, it can also introduce bias if the regression model is misspecified.
```

```python
#Therefore normalization we can apply removal of outliers by replacing it bu mode
```

```python
from numpy import percentile
l=list(df.columns)
for i in l:
    lst=df[i].to_list()
    Q1,Q2,Q3 = percentile(df[i], [25, 50, 75])
    IQR=Q3-Q1
    LOWER_FENCE=Q1-(1.5*IQR)
    HIGHER_FENCE=Q3+(1.5*IQR)
    for j in range(len(lst)):
        if (lst[j]>HIGHER_FENCE) or (lst[j]<LOWER_FENCE):
            lst[j]=df[i].mode()[0]
    df[i]=lst
# df['Last Updated'] =df['Last Updated'].replace('1.0.19',df['Last Updated'].mode()[0])
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(l)):
    plt.subplot(4, 3, i+1)
    sns.boxplot(df[l[i]])
    plt.xlabel(l[i])
    plt.tight_layout()
```
